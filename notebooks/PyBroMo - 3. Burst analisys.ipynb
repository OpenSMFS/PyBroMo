Specify here the parameters of the simulation to load:

```python
d_EM_kHz = 20.
d_BG_kHz = 6
a_EM_kHz = 180.
a_BG_kHz = 3
FRET_val = 100.*a_EM_kHz/(a_EM_kHz+d_EM_kHz)
print "Simulated FRET value: %.1f%%" % FRET_val
```

```python
# These are used for the file name
ID = '1+2+3+4+5+6'
t_tot = '480' # string without trailing 's'
num_p = '30'
pM = '64'

# These are metadata associated to the timestamps
t_step = 0.5e-6
clk_p = t_step/32. # with t_step=0.5us -> 156.25 ns
```

Here we compute the file name from the parameters:

```python
d_EM_kHz_str = "%04d" % d_EM_kHz
a_EM_kHz_str = "%04d" % a_EM_kHz
d_BG_kHz_str = "%04.1f" % d_BG_kHz
a_BG_kHz_str = "%04.1f" % a_BG_kHz

print "D: EM %s BG %s "%(d_EM_kHz_str,d_BG_kHz_str)
print "A: EM %s BG %s "%(a_EM_kHz_str,a_BG_kHz_str)
```

```python
fname_d = "ph_times_{t_tot}s_D1.2e-11_{np}P_{pM}pM_step0.5us_ID{ID}_EM{em}kHz_BG{bg}kHz.npy".format(em=d_EM_kHz_str, bg=d_BG_kHz_str, t_tot=t_tot, pM=pM, np=num_p, ID=ID)
fname_a = "ph_times_{t_tot}s_D1.2e-11_{np}P_{pM}pM_step0.5us_ID{ID}_EM{em}kHz_BG{bg}kHz.npy".format(em=a_EM_kHz_str, bg=a_BG_kHz_str, t_tot=t_tot, pM=pM, np=num_p, ID=ID)
print fname_d
print fname_a
```

Load the timestamps arrays:

```python
ph_times_d = np.load(SIM_PH_DIR+fname_d)
ph_times_a = np.load(SIM_PH_DIR+fname_a)
```

```python
ph_times, a_em = merge_DA_ph_times(ph_times_d, ph_times_a)
ph_times_int = (ph_times/clk_p).astype('int64')
```

# FRET fitting

```python
NN = r_[100:1001:100,1100:3001:100]
E_fit_m = zeros(NN.size)
E_fit_slope = zeros(NN.size)
E_fit_hist = zeros(NN.size)
E_fit_poiss = zeros(NN.size)
E_fit_poisst = zeros(NN.size)
E_fit_binom = zeros(NN.size)
    
for i,N in enumerate(NN):
    ds = Sel(d, select_bursts_topN_nda, N=N)
    
    E_fit_m[i] = ds.fit_E_m(weights='size')
    
    E_fit_slope[i] = ds.fit_E_minimize(kind='slope', weights='inv_sqrt')
    
    E_fit_hist[i] = ds.fit_E_generic(fit_fun=gaussian_fit_hist, weights='size')
    
    E_fit_poiss[i] = ds.fit_E_ML_poiss(method=1)
    E_fit_poisst[i] = ds.fit_E_ML_poiss(method=2)
    
    E_fit_binom[i] = ds.fit_E_ML_binom()
```

```python
figure()
fill_between(NN, ones(NN.size)*(FRET_val-1.), ones(NN.size)*(FRET_val+1.), alpha=0.1)
fill_between(NN, ones(NN.size)*(FRET_val-.5), ones(NN.size)*(FRET_val+.5), alpha=0.1)

plot(NN, 100*E_fit_m, label='W. Mean')
plot(NN, 100*E_fit_slope, label='Slope')
plot(NN, 100*E_fit_hist, label='W. Hist')
plot(NN, 100*E_fit_poiss, label='ML Poiss na')
plot(NN, 100*E_fit_poisst, label='ML Poiss nt')
plot(NN, 100*E_fit_binom, label='ML Binom')
legend(loc='best', ncol=2, frameon=False)
axhline(FRET_val, lw=2, color='k')
title(d.Name().replace('BS_', '\nBS_')); ylim(88,92);
```
